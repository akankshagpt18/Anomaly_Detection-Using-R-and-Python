```python
import sqlite3
from sqlite3 import Error
import pandas as pd
import csv
```

```python
def create_connection(sqlite_db):
    """ create a database connection to the SQLite database
        specified by the sqlite_db
    :param sqlite_db: database file
    :return: Connection object or None
    """
    try:
        conn = sqlite3.connect(sqlite_db)
        return conn
    except Error as e:
        print(e)
    
    
    return None
```

```python
def select_tables(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    cur.execute("Select svid from sat_data")
    
    rows = cur.fetchall()
    for row in rows:
        print(row)
    
        
```

```python
                                
def main():
    database= "TORGI-GNSS-20180826T060733.348Z.gpkg"
    
    #create database connection
    conn = create_connection(database)
    with conn:
        print("Select tables")
        select_tables(conn)
        
        
        
if __name__ == '__main__':
  main() 
```
