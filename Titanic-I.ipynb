```python
def male_female_child(passenger):
    age,sex=passenger
    
    if age<16:
        return 'child'
    else:
        return sex
```

```python
titanic['person']=titanic[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
titanic[0:10]
```

```python
sns.factorplot('Pclass', data=titanic, hue='person', kind='count')
```

```python
titanic['Age'].hist(bins=70)
```

```python
fig = sns.FacetGrid(titanic, hue='Sex', aspect =5)
fig.map(sns.kdeplot, 'Age',shade=True)
oldest=titanic['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic, hue='person', aspect =4)
fig.map(sns.kdeplot, 'Age',shade=True)
oldest=titanic['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic, hue='Pclass', aspect =5)
fig.map(sns.kdeplot, 'Age',shade=True)
oldest=titanic['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```
